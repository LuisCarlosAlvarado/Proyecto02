#                                            Portfolio Allocation

Se importan los datos de los últimos 3 años de las acciones  de **Apple, Disney, Tesla** y **Nike**. Para esto se utiliza Datareader, obteniendo la información de Yahoo Finance. Únicamente queremos el Adj. Close, porque nos interesa el movimiento del precio de las acciones sin tomar en cuenta los dividendos que pagaron las empresas a sus accionistas.

```python
def get_data(name,start,end):
    return web.DataReader(name,'yahoo','2016-01-01','2019-01-01')['Adj Close']
```

```python
name='AAPl'
apple=pd.DataFrame(columns=['Adj Close'])
apple['Adj Close']= get_data(name,'2016-01-01','2019-01-01')
```

```python
name='DIS'
disney=pd.DataFrame(columns=['Adj Close'])
disney['Adj Close']= get_data(name,'2016-01-01','2019-01-01')
```

```python
name='TSLA'
tesla=pd.DataFrame(columns=['Adj Close'])
tesla['Adj Close']= get_data(name,'2016-01-01','2019-01-01')
```

```python
name='NKE'
nike=pd.DataFrame(columns=['Adj Close'])
nike['Adj Close']= get_data(name,'2016-01-01','2019-01-01')
```

Se define una nueva función, que es el rendimiento, definido como el precio de la acción en el día dividido entre el precio inicial.

Se asigna un porcentaje de nuestro $100{%}$, determinado por nosotros a cada empresa. 

Al final se multiplica el retorno diario por el  porcentaje.

```python
for stock_df in (apple,disney,tesla,nike):
    stock_df["Normed Return"]=stock_df["Adj Close"]/stock_df.iloc[0]["Adj Close"]

for stock_df, allo in zip((apple,disney,tesla,nike),[.3,.2,.4,.1]):
    stock_df["Allocation"]=stock_df["Normed Return"]*allo
    
for stock_df in (apple,disney,tesla,nike):
    stock_df["Position Values"]=stock_df["Allocation"]*1000000

apple.head()
```

Lo que se hace en esta parte es formar el portafolio, aquí se van a concatenar los retornos de las 4 empresas multiplicados por
sus asignaciones. Esto quedará en una **Total Position**

```python
all_pos_vals=[apple["Position Values"],disney["Position Values"],
             tesla["Position Values"],nike["Position Values"]]

portfolio_val=pd.concat(all_pos_vals,axis=1)
portfolio_val.columns=["Apple Pos","Disney Pos","Tesla Pos","Nike Pos"]
portfolio_val["Total Pos"]=portfolio_val.sum(axis=1)
portfolio_val.head()
```

Mostramos un gráfico de cómo se movió el rendimiento del **PORTAFOLIO** durante el $2019$.

```python
portfolio_val["Total Pos"].plot(figsize=(10,8))
plt.title("Total Portfolio Value")
```

Mostramos un gráfico de cómo se movió el rendimiento de las **ACCIONES** durante el $2019$

```python
portfolio_val.drop("Total Pos",axis=1).plot(figsize=(10,8))
plt.title("Stock's Positions")
```

## Cumulative return 

El retorno acumulado se obtiene dividiendo el valor del portafolio al final del periodo entre la inversión inicial.

```python
cumulative_return=100*(portfolio_val["Total Pos"][-1]/portfolio_val["Total Pos"][0]-1)
cumulative_return
```

## Sharpe Ratio

La función Sharpe Ratio es la más importante en esta parte del proyecto, es una combinación del riesgo que tienen las acciones relacionadas al retorno que genera cada una de ellas. El riesgo tiene que ver con la volatilidad de las acciones, es decir, la estabilidad que tienen, que tanto se mueven **a la alza** o **a la baja** durante el periodo. 

La obtenemos dividiendo la media del retorno diario del **PORTAFOLIO**, entre su desviación estándar.

```python
portfolio_val["Daily Return"]=portfolio_val["Total Pos"].pct_change(1)
portfolio_val.head()
```

```python
sharpe_ratio=portfolio_val["Daily Return"].mean()/portfolio_val["Daily Return"].std()
sharpe_ratio
```

```python
annualized_sharpe_ratio=(252**0.5)*sharpe_ratio
annualized_sharpe_ratio
```

# Portfolio Optimization

Aquí buscamos obtener el mayor rendimiento posible combinado con la menor volatilidad, es decir, mayor rendimiento menor riesgo, completamente contrario a la lógica de los mercados.

```python
def get_data(name,start,end):
    return web.DataReader(name,'yahoo','2016-01-01','2019-01-01')['Adj Close']
```

```python
name='AAPl'
apple=pd.DataFrame(columns=['Adj Close'])
apple['Adj Close']= get_data(name,'2016-01-01','2019-01-01')
```

```python
name='DIS'
disney=pd.DataFrame(columns=['Adj Close'])
disney['Adj Close']= get_data(name,'2016-01-01','2019-01-01')
```

```python
name='TSLA'
tesla=pd.DataFrame(columns=['Adj Close'])
tesla['Adj Close']= get_data(name,'2016-01-01','2019-01-01')
```

```python
name='NKE'
nike=pd.DataFrame(columns=['Adj Close'])
nike['Adj Close']= get_data(name,'2016-01-01','2019-01-01')
```

Volvemos a pedir el precio de las acciones durante los últimos 3 años, como lo hicimos anteriormente. Mostramos los datos en una sola tabla para trabajar con ellos al mismo tiempo.

```python
stocks=pd.concat([apple,disney,tesla,nike],axis=1)
stocks.columns=["Apple","Disney","Tesla","Nike"]
stocks.head()
```

Utilizamos la función del retorno logarítmico, que es muy cercano al rendimiento que ya habíamos calculado.

Mostramos también sus histogramas para ver cómo se comportan los datos, lo esperado es ver una campana de Gauss en cada acción.

```python
log_ret=np.log(stocks/stocks.shift()).dropna()
log_ret.head()
```

```python
log_ret.mean()
```

```python
log_ret.hist(bins=100,figsize=(12,8))
plt.tight_layout()
```

```python
log_ret.cov()*252
```

Lo siguiente es definir los parámetros que necesitamos para poder graficar. Se supone que tenemos 3 variables importantes, volatilidad, retorno y Sharpe Ratio, pero no lo graficaremos en 3D, utilizarmos las 3 variabes en un gráfico 2D.

Todo depende de las dimensiones que le demos al arreglo y los puntos aleatorios que le demos.

```python
np.random.seed(101)
num_ports=5000
all_weights=np.zeros((num_ports,len(stocks.columns)))
ret_arr=np.zeros(num_ports)
vol_arr=np.zeros(num_ports)
sharpe_arr=np.zeros(num_ports)

for i in range(num_ports):
    #Weights
    weights=np.array(np.random.random(4))
    weights=weights/np.sum(weights)

    #Save weights
    all_weights[i,:]=weights
    
    #Expected Return
    ret_arr[i]=np.sum((log_ret.mean()*weights)*252)

    #Expected volatility
    vol_arr[i]=np.sqrt(np.dot(weights.T,np.dot(log_ret.cov()*252,weights)))

    #Sharpe Ratio
    sharpe_arr[i]=ret_arr[i]/vol_arr[i]
```

```python
sharpe_arr.max()
```

```python
sharpe_arr.argmax()
```

## Investment Portfolio

Usando el valor máximo en el arreglo, definimos anteriormente cuál será la volatilidad y rendimiento en base al valor máximo.

Con esos datos previamente calculados grafiamos en 2D. Para la 3er variable utilizaremos un contraste de colores.

```python
#COLORS PLASMA; INFERNO, MAGMA, CIVIDIS, VIRIDIS, COOL
plt.figure(figsize=(12,8))
plt.scatter(vol_arr,ret_arr,c=sharpe_arr,cmap="cool")
plt.colorbar(label="Sharpe Ratio")
plt.xlabel("Volatility")
plt.ylabel("Return")
plt.title("Investment Portfolio")

plt.scatter(max_sr_vol,max_sr_ret,c="black",s=100)
```

## Simulación para portafolios

Ahora emplearemos lo visto en clase para una simulación  de 1000 **PORTAFOLIOS**, es decir, juntando las 4 acciones seleccionadas.

```python
R=np.log(portfolio_val['Total Pos']/portfolio_val['Total Pos'].shift()).dropna()
```

```python
def simR(mu,std,start,ndays,N):
    dates=pd.date_range(start,periods=ndays)
    data=np.random.normal(mu,std,(ndays,N))
    return pd.DataFrame(index=dates,data=data)
```

```python
datosR=simR(R.mean(),R.std(),'2019-01-01',365,1000)
```

```python
datosR.plot(grid=True,legend=False)
```

Basados en la simulación de los rendimientos, se muestra el gráfico del movimiento total del **PORTAFOLIO** hasta 2020.

```python
simCloses=portfolio_val["Total Pos"].iloc[-1]*((1+datosR).cumprod())
```

```python
simCloses.plot(grid=True,legend=False)
```

Calculamos el Umbral, nosotros definimos en rendimiento esperado del 20%, que es bastante bajo considerando las 4 acciones que tenemos, pero es un buen rendimiento promedio.

```python
U=portfolio_val["Total Pos"].iloc[-1]*1.2
dates=pd.date_range('2019-01-01',periods=365)
Umbral=pd.DataFrame(index=dates,columns=['U'],data=U*np.ones(len(dates)))
Umbral
```

```python
simCloses_U=pd.concat([simCloses,Umbral],axis=1)
simCloses_U.plot(legend=False,grid=True)
```

Este es el rendimiento que probablemente hubiésemos tenido si nuestro portafolio a inicios de 2019 tuviese las acciones que escogimos, pasando del 35% de rendimiento.

```python
C=simCloses>U
P=pd.DataFrame(np.mean(C,axis=1))
P.plot(legend=False,grid=True,figsize=(10,6))
```
